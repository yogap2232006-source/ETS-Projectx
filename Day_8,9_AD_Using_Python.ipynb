```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.stats import ttest_1samp

data = [
    {'Month': 'Jan', 'Sales': 12000, 'Spend': 3000, 'Visitors': 1200},
    {'Month': 'Feb', 'Sales': 15000, 'Spend': 3500, 'Visitors': 1350},
    {'Month': 'Mar', 'Sales': 17000, 'Spend': 4000, 'Visitors': 1600},
    {'Month': 'Apr', 'Sales': 14000, 'Spend': 3200, 'Visitors': 1280},
    {'Month': 'May', 'Sales': 18000, 'Spend': 4200, 'Visitors': 1700},
]
df = pd.DataFrame(data)
df
df.describe()
df.corr(numeric_only=True)
X = df[['Spend']]
y = df['Sales']
model = LinearRegression().fit(X,y)
model.predict([[5000]])
coef = model.coef_[0]
intercept = model.intercept_
(20000 - intercept) / coef
plt.plot(df['Month'], df['Sales'])
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.stats import ttest_1samp

data = [
    {'Month': 'Jan', 'Sales': 12000, 'Spend': 3000, 'Visitors': 1200},
    {'Month': 'Feb', 'Sales': 15000, 'Spend': 3500, 'Visitors': 1350},
    {'Month': 'Mar', 'Sales': 17000, 'Spend': 4000, 'Visitors': 1600},
    {'Month': 'Apr', 'Sales': 14000, 'Spend': 3200, 'Visitors': 1280},
    {'Month': 'May', 'Sales': 18000, 'Spend': 4200, 'Visitors': 1700},
]
df = pd.DataFrame(data)
df
df.describe()
df.corr(numeric_only=True)
X = df[['Spend']]
y = df['Sales']
model = LinearRegression().fit(X,y)
model.predict([[5000]])
coef = model.coef_[0]
intercept = model.intercept_
(20000 - intercept) / coef
plt.plot(df['Month'], df['Sales'])
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
t_stat, p_val = ttest_1samp(df['Sales'], 15000)
t_stat, p_val
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("salesdata.csv")
df.head()
print(df.shape)
print(df.columns)
df.describe()
df.isnull().sum()
df['Date'] = pd.to_datetime(df['Date'])
df.drop_duplicates(inplace=True)
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df['Revenue'] = df['Quantity'] * df['Price']
monthly_sales = df.groupby('Month')['Revenue'].sum()
monthly_sales
plt.figure(figsize=(10,5))
df.groupby('Month')['Revenue'].sum().plot(marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()
```
